```python
#torch.mm == torch.matmul

tensor_a = torch.tensor([[1,2],
                         [3,4],
                         [5,6]])
tensor_b = torch.tensor([[7,10],
                         [8,11],
                         [9,12]])

torch.mm(tensor_a, tensor_b)
```

```python
## We can fix this using taking a transpose

tensor_b.T
```

```python
torch.mm(tensor_a, tensor_b.T)
```

```python
#Reshaping, Stacking, Squeezing,unsqueezing

x= torch.arange(1.,10.)
x,x.shape
```

```python
x_reshaped = x.reshape(1,9)
x_reshaped
```

```python
x_reshaped = x.reshape(9,1)
x_reshaped
```

```python
x_reshaped = x.reshape(1,3,3)
x_reshaped
```

```python
#view

z = x.view(1,9)
#Chaning z changes x cuz here z is directly changing the memory of x(same pointers)
```

```python
z[:,0] = 5
z,x
```

```python
#stack tensors

x_stacked = torch.stack([x,x,x,x],dim= 1)
x_stacked
```
