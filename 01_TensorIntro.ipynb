```python
## Tensor DataType

#Imp cuz 3 errors u'll run into in Pytorch or Deep Learning are
# 1. Tensor not right Datatype
#2. Tensor not right shape
#3. Tensor not on right device

float_32_tensor = torch.tensor([3.0,5.0,7.0],
                                dtype = None, #What datatype is the tensor
                                device = None, #If we want to use GPU for this tensor
                                requires_grad = False # If we want to track the gradient for our tensor(for backpropagation i'm guessing)
                              )
float_32_tensor
```

```python
float_16_tensor = float_32_tensor.type(torch.float16)
float_16_tensor
```
